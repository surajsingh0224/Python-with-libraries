```python
# ADVANCE ARRAY  MANUPLATION
```

```python
# Stacking the array
```

```python
A = np.array([[1,2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
```

```python
horizontal_stack = np.hstack((A, B))
horizontal_stack
```

```python
vertical_stack = np.vstack((A, B))
vertical_stack
```

```python
col_stack = np.column_stack((A, B))
col_stack
```

```python
# SPLITTING ARRAY
```

```python
vertical_split = np.vsplit(c, 4)
```
