```python
from tkinter import *
```

```python
win = Tk()

win.title("Layout management")
win.config(bg = "yellow")

lab = Label(win, text = "Suraj Singh", font = ("Time New Roman", 30, "bold"), bg = "Red")
# lab.pack(padx=20, pady=200)
# lab.pack(ipadx=100, ipady = 100)
# lab.pack(ipadx=100, fill="x")
# lab.pack(ipadx=100, ipady = 100, expand=True)
lab.pack()
win.mainloop()
```

```python
# GRID
win = Tk()

win.title("Layout management")
win.config(bg = "yellow")

lab = Label(win, text = "Suraj Singh", font = ("Time New Roman", 30, "bold"), bg = "Red")
lab.grid(row = 0, column= 0, padx=20, pady=20, columnspan=2)


win.mainloop()
```

```python
# PLACE
win = Tk()

win.title("Layout management")
win.config(bg = "yellow")

lab = Label(win, text = "Suraj Singh", font = ("Time New Roman", 30, "bold"), bg = "Red")
lab.place(x = 200, y = 200, height=100, width=200)


win.mainloop()
```
