```python
import matplotlib.pyplot as plt
import seaborn as sns
```

We use distribution plots for visualize the distribution of quantative data. 
 some plots that allow us to visualize the distribution of a data sets. They Are : 

1. distplot / histplot

```python
# Loading the dataset from the internet 
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# Plotting of total_bills and tips in the subplot form using matplotlib 
plt.subplot(1, 2, 1)
sns.histplot(df['total_bill'], bins= 20, kde = True)
plt.subplot(1, 2, 2)
sns.histplot(df['tip'], bins = 20, kde = True)
```

2. joint plot

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter', color='darkcyan')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data= df, kind= 'reg')
```

3. pair plot 

```python
sns.pairplot(df, hue = 'size', palette='bright')
```

4. rug plot 

```python
sns.rugplot(df['tip'])
```
