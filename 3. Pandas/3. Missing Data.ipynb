```python
import numpy as np
import pandas as pd
```

```python
# FINDING MISSING DATA
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() # TRUE MEANS THIS IS NAN. 
```

```python
df.isna().sum()
```

```python
df.isna().any() # whether that column contains at least one missing (NaN or None) value. 
```

```python
# REMOVING A MISSING DATA
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(thresh= 1) #  will keep all rows in the DataFrame that have at least one non-missing (non-NaN) value. 
```

```python
# FILLING THE MISSING DATA
```

```python
df
```

```python
df.fillna(0)
```

```python
values = {'A': 0,
          'B': 100,
          'C': 300,
          'D': 400
         }

df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
